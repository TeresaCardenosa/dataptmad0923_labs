```python
# imports
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
```

```python
data = pd.read_csv('./iris-data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
# 5 nulos en la columna petal_leght_cm >> mejor cargase esas filas 
```

```python
data = data.dropna()
```

```python
data.info()
```

```python
# Columna class es object > Transformar a n√∫mero
```

```python
data['class'].value_counts()
```

```python
'''
Homogeneizar valores:
  - Iris-virginica (50)
  - Iris-versicolor (45) == versicolor (5)
  - Iris-setosa (44) == Iris-setossa (1)

'''
```

```python
data['class'] = data['class'].replace(['versicolor', 'Iris-setossa'], ['Iris-versicolor', 'Iris-setosa'])
```

```python
data['class'].value_counts()
```

```python
data.head()
```

```python
# Now: One-hot encoding para la colum de class
```

```python
iris_one_hot_encoding = pd.get_dummies(data['class'], drop_first=True, dtype=float)
iris_one_hot_encoding
```

```python
# Concat de los dos dataframes y eliminar la columna original de class
```

```python
df_prep = pd.concat([data, iris_one_hot_encoding], axis=1)    # Concatear los dos DF a√±adiendo columnas
```

```python
df_prep
```

```python
df_prep = df_prep.drop(['class'], axis=1)            # Eliminar la columna class 
```
